# Récupérer les données de LMVH

```python
import yfinance as yf
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
import numpy as np
from sklearn.metrics import mean_squared_error, mean_absolute_error
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
import pandas as pd
```

```python
# S&P500 a pour ticker "^GSPC"
# Récupérer les données financières sur yfinance
data = yf.download("^GSPC", start="2007-01-01", end="2025-01-01")
print(data.head())

# Afficher prix de clôture, et volume
data['Close'].plot(title="Prix de clôture S&P500 (2007-2025)")
plt.show()

data['Volume'].plot(title="Volume S&P500  (2007-2025)")
plt.show()
```

# Normaliser les données financières

Les réseaux de neurones fonctionnent si les données sont dans une plage restreinte (entre 0,1 ou -1,1).

=> X = (X - Xmin) / (Xmax - Xmin)

```python
close_prices = data["Close"].values.reshape(-1,1)
scaler = MinMaxScaler(feature_range=(0,1))
close_prices_scaled = scaler.fit_transform(close_prices)
print(close_prices_scaled)
```

# Séquences temporelles

```python
def create_sequences(data, seq_length):
    X = []
    y = []
    for i in range(len(data) - seq_length):
        X.append(data[i:i+seq_length])
        y.append(data[i+seq_length])
    return np.array(X), np.array(y)

SEQ_LENGTH = 60 
X, y = create_sequences(close_prices_scaled, SEQ_LENGTH)

print(f"X shape: {X.shape}")  
print(f"y shape: {y.shape}")  
```

# LSTM (long short term memory)



```python
# X doit avoir la forme (samples, seq_length, features)
print(f"Shape de X: {X.shape}")  # ex: (nombre_exemples, 60, 1)

# 3. Construire le modèle
model = Sequential()
model.add(LSTM(50, return_sequences=False, input_shape=(X.shape[1], X.shape[2]))) 
# 50 unités LSTM, return_sequences=False car on veut une sortie finale
model.add(Dense(1))  # couche dense finale pour prédire le prix unique

# 4. Compiler le modèle
model.compile(optimizer='adam', loss='mean_squared_error')

# 5. Entraîner le modèle
history = model.fit(X, y, epochs=20, batch_size=32, validation_split=0.2)

# Le modèle est prêt à prédire !
```

```python
# Train/ test split
train_size = int(len(X) * 0.8) # 80% train, 20% test
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

# Entraîner le modèle
model.fit(X_train, y_train, epochs=20, batch_size=32, validation_data=(X_test, y_test))

# Prédiction 
y_pred_scaled = model.predict(X_test)

# Inverser la normalisation
y_pred = scaler.inverse_transform(y_pred_scaled)
y_test_real = scaler.inverse_transform(y_test)

# Visualiser la prédiction vs réaliter
dates = data.index.values 
dates_test = dates[len(dates) - len(y_test):] 
plt.figure(figsize=(12,6))
plt.plot(dates_test, y_test_real, label='Prix réel')
plt.plot(dates_test, y_pred, label='Prix prédit')
plt.title("Prédiction des prix LVMH")
plt.xlabel("Date")
plt.ylabel("Prix")
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

```python
# y_test_real et y_pred : les vraies valeurs et les prédictions à l’échelle originale
rmse = np.sqrt(mean_squared_error(y_test_real, y_pred))
mae = mean_absolute_error(y_test_real, y_pred)

print(f"Root Mean Squared Error (RMSE) : {rmse:.4f}")
print(f"Mean Absolute Error (MAE) : {mae:.4f}")
```

```python
last_sequence = X_test[-1] 

def predict_future_days(model, last_seq, n_days):
    preds = []
    current_seq = last_seq.copy()

    for _ in range(n_days):
        pred = model.predict(current_seq.reshape(1, current_seq.shape[0], 1))
        preds.append(pred[0,0])
        current_seq = np.append(current_seq[1:], pred[0,0])
    return np.array(preds)

n_days_to_predict = 30
future_preds_scaled = predict_future_days(model, last_sequence, n_days_to_predict)

future_preds = scaler.inverse_transform(future_preds_scaled.reshape(-1,1))

last_date = data.index[-1]
future_dates = pd.date_range(start=last_date + pd.Timedelta(days=1), periods=n_days_to_predict, freq='B')

plt.figure(figsize=(12,6))
plt.plot(data.index, data['Close'], label="Prix réel")
plt.plot(future_dates, future_preds, label="Prédiction future")
plt.title("Prédiction des prix futurs S&P500")
plt.xlabel("Date")
plt.ylabel("Prix")
plt.legend()
plt.show()
```
